```python
####interquartile range###
####outlier--->.an outlier is an observation that lies on abnormal distance from other values in a random sample##
###impact --->dissort the mean,increase variance

###outlier detection and handling
q1 = df["horsepower"].quantile(0.25)
q3 = df["horsepower"].quantile(0.75)

iqr = q3 -q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

outliers = df[(df["horsepower"] < lower) | ((df["horsepower"] > upper))]

print(outliers[["make","horsepower"]])
```

```python
#####z_score method ###
from scipy import stats

df["zscore"] = stats.zscore(df["horsepower"])
z_outliers = df[np.abs(df["zscore"]) > 3]
print(z_outliers[["make","horsepower"]])
```

```python
###label encoding
cy1_map = {"two":2,"four":4}
df["num-of-doors_encoded"] = df["num-of-doors"].map(cy1_map)
print(df[["num-of-doors","num-of-doors_encoded"]])

unique = pd.unique(df["num-of-doors"])
print(unique)
```

```python
unique = pd.unique(df["make"])
print(unique)
len(unique)
make_list = {"alfa-romero":1,"audi":2,"bmw":3,"chevrolet":4,"dodge":5,"honda":6,"isuzu":7,"jaguar":8,"mazda":9,"mercedes-benz":10}
df["make_encoded"] = df["make"].map(make_list)
print(df[["make","make_encoded"]])
```

```python
####data scaling (normalization)/min-max standardization
##min-max standard[0,1]
unique = pd.unique(df["wheel-base"])
print(unique)
len(unique)

df["wheel-base_min_max"] = (df["wheel-base"] - df["wheel-base"].min()) / (df["wheel-base"].max() - df["wheel-base"].min())
print(df["wheel-base_min_max"])
```

```python
df["price_min_max"] = (df["price"] - df["price"].min()) / (df["price"].max() - df["price"].min())
print(df["price_min_max"])
```

```python
####standardization (mean=0,std=1)
df["wheel-base_std"] = (df["wheel-base"] - df["wheel-base"].mean()) / df["wheel-base"].std()
print(df["wheel-base_std"])
```

```python
df["price_std"] = (df["price"] - df["price"].mean()) / df["price"].std()
print(df["price_std"])
```

```python
###feature selection####
####feature engineering####
# 1 feature selection(correlation)

corr_matrix = df.corr(numeric_only=True)

print(corr_matrix["price"].sort_values(ascending=False))
```

```python
#####avrage mpg 
df["avg-mpg"] = (df["city-mpg"] + df["highway-mpg"]) / 2
print(df["avg-mpg"])
```

```python
df["power_of_weight"] = (df["curb-weight"] + df["horsepower"]) / 2
print(df["power_of_weight"])
```

```python
#####  HANDLING IMBALANCE DATA (SMOTE--->OVERSAMPLING)
mean_price = df["price"].mean()

df["high_price"] = np.where(df["price"] > mean_price , 1 , 0)
print(df["high_price"].value_counts())
```

```python
###reducing data majority class (under sampling)
majority = df[df["high_price"] == 0]
majority_1 = df[df["high_price"] == 1]

majority_under = majority.sample(len(majority_1))
df_under = pd.concat([majority_under,majority_1],axis = 0)
print(df_under["high_price"].value_counts())
                                 
```

```python
####over sampling####
majority_over = majority_1.sample(len(majority),replace=True)
df_over = pd.concat([majority_over,majority],axis = 0)
print(df_over["high_price"].value_counts())
```
